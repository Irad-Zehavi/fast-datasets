```python
#| export
from fastai.vision.all import *

import fastai_datasets.patches
```

```python
#| export
def CIFAR10():
    return DataBlock(
        blocks=(ImageBlock, CategoryBlock),
        get_items=get_image_files,
        get_y=parent_label,
        splitter=GrandparentSplitter('train', 'test'),
    ).datasets(untar_data(URLs.CIFAR))
```

```python
CIFAR10().dls().show_batch()
```

```python
#| export
def CIFAR100():
    return DataBlock(
        blocks=(ImageBlock, CategoryBlock),
        get_items=get_image_files,
        get_y=parent_label,
        splitter=FuncSplitter(lambda p: p.parents[2].name == 'test'),
    ).datasets(untar_data(URLs.CIFAR_100))
```

```python
CIFAR100().dls().show_batch()
```
